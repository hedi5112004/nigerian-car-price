```python
import pandas as pd
data = pd.read_csv(r"C:\Users\Hedi\Downloads\Nigerian_Car_Prices (1).csv")
data
```

```python
 data.describe()
```

```python
data.info()
```

```python
data['Price'].replace(to_replace=',', value='', regex=True, inplace=True)
data["Price"] = data["Price"].astype("float64")
data.info()
```

```python
data.isnull().sum()
```

```python
data["Year of manufacture"].fillna(data["Year of manufacture"].mean(),inplace=True)
data["Condition"].fillna("G6",inplace=True)
data["Mileage"].fillna(data["Mileage"].mean(),inplace=True)
data["Engine Size"].fillna(data["Engine Size"].mean(),inplace=True)
data["Fuel"].fillna("G6",inplace=True)
data["Transmission"].fillna("G6",inplace=True)
data["Build"].fillna("G6",inplace=True)
```

```python
data.isnull().sum()
```

```python
import seaborn as sns
sns.displot(data['Year of manufacture'])
```

```python
sns.displot(data['Mileage'])
```

```python
q = data['Mileage'].quantile(0.99)
data = data[data['Mileage']<q]
```

```python
sns.displot(data['Mileage'])
```

Certainly, here's an alternative expression:
This adjustment has notably improved the distribution, now resembling a normal distribution.

```python
sns.displot(x=data['Engine Size'])
```

```python
data = data[data['Engine Size']<8382]
data = data[data['Engine Size']>624]
sns.displot(x=data['Engine Size'])
```

```python
import matplotlib.pyplot as plt
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize =(15,3))
ax1.scatter(data['Year of manufacture'],data['Price'])
ax1.set_title('Price and Year')
ax2.scatter(data['Engine Size'],data['Price'])
ax2.set_title('Price and Engine Size')
ax3.scatter(data['Mileage'],data['Price'])
ax3.set_title('Price and Mileage')
plt.show()
```

```python
import numpy as np
log_price = np.log(data['Price'])
data['log_price'] = log_price
data
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize =(15,3)) 
ax1.scatter(data['Year of manufacture'],data['log_price'])
ax1.set_title('Price and Year')
ax2.scatter(data['Engine Size'],data['log_price'])
ax2.set_title('Price and Engine Size')
ax3.scatter(data['Mileage'],data['log_price'])
ax3.set_title('Price and Mileage')
plt.show()
```
